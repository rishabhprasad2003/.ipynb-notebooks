```python
overs = [0,4,8,12,16,20]
team1 = [0,30,67,98,123,145]
team2 = [0,27,75,103,110,135]
plt.plot(overs,team1,label = 'Team1',marker = 'd')
plt.plot(overs,team2,label = 'Team2',marker = '*')
plt.title('ANALYSIS OF CRICKET MATCH',fontsize = 20)
plt.xlabel('OVERS')
plt.ylabel('RUNS')
plt.legend()
```

```python
#SPORTS WATCH DATA
x = np.array([60,70,80,90,100])
y = [250,270,260,280,300]
plt.plot(x,y)
plt.title('SPORTS WATCH DATA',fontsize = 20)
plt.xlabel('Average Pulse')
plt.ylabel('Calorie Burnage')
plt.grid()# will create grid lines for me
```

```python
#SUB PLOT
fig = plt.figure()

x1 = [1,2,3,4]
x2 = [5,6,7,8]
y1 = np.random.randint(0,10,4)
y2 = np.random.randint(0,10,4)

ax1 = fig.add_subplot(331)
ax2 = fig.add_subplot(338)
ax3 = fig.add_subplot(339)

ax1.plot(x1,y1)
ax2.plot(x2,y2)
```
